### Getting all stocks from IBOV, straight from B3

```python

url = 'https://sistemaswebb3-listados.b3.com.br/indexProxy/indexCall/GetPortfolioDay/eyJpbmRleCI6IklCT1YiLCJsYW5ndWFnZSI6InB0LWJyIn0='
# lendo conteúdo json da url
r = requests.get(url)
b3 = json_normalize(r.json(), 'results')['cod'].tolist()
# acrescentando ".SA" ao final de cada ticker
b3_tickers = [ticker + '.SA' for ticker in b3]
# adding ^BVSP (Ibovespa) to the list
b3_tickers.append('^BVSP')
b3_tickers
```

### Downloading all tickers using yfinance

```python
all = yf.download(b3_tickers)['Adj Close'].reset_index().rename(columns={'^BVSP':'IBOV'})
all.columns = [ticker.replace('.SA', '') for ticker in all.columns]
all
```

### Calculating Returns

```python
returns = all.copy()
returns.iloc[:, 1:] = all.iloc[:, 1:].pct_change() # pct returns from all columns except de date
returns
```

### Calculating Beta

```python
# Calculating beta from all stocks in "all", using the column "IBOV" as the market
# beta = cov(stock, market) / var(market)

all_beta = pd.DataFrame()

# Iterating trough all columns, except the first and the last one (date and IBOV)
for ticker in all.columns[1:-1]:
    # Dropping NaN values from the current ticker and IBOV (so the stock gets the same number of rows as the market)
    current_ticker = returns[[ticker, 'IBOV']].dropna(subset=ticker)
    # Calculating beta and appending to the dataframe
    all_beta = all_beta.append([[ticker, current_ticker[ticker].cov(current_ticker['IBOV']) / current_ticker['IBOV'].var()]], ignore_index=True)

all_beta.columns = ['Ticker', 'Beta'] # Renaming column names
# Classifying beta
all_beta['type'] = np.select([all_beta['Beta'] <= 0.8, all_beta['Beta'] >= 1.2], ['Defensive', 'Aggressive'], default='Neutral')

all_beta
```
